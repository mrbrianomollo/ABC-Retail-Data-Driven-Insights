```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
customer_sales = pd.read_csv("customer_sales_data.csv")
inventory = pd.read_csv("inventory_data.csv")

```

```python
# Analysis 1: Top Customers
top_customers = customer_sales.groupby("CustomerName")["SaleAmount"].sum().sort_values(ascending=False).head(10)
print("Top Customers:\n", top_customers)

# Visualization of Top Customers (Horizontal Bar Plot)
plt.figure(figsize=(10, 6))
sns.barplot(y=top_customers.index, x=top_customers.values, orient='h')
plt.title("Top Customers by Sale Amount")
plt.xlabel("Total Sale Amount")
plt.ylabel("Customer Name")
plt.show()
```

```python
# Analysis 2: Top Selling Products
top_products = customer_sales.groupby("ProductName")["SaleAmount"].sum().sort_values(ascending=False).head(10)
print("\nTop Selling Products:\n", top_products)

# Visualization of Top Selling Products (Pie Chart)
plt.figure(figsize=(8, 8))
plt.pie(top_products, labels=top_products.index, autopct='%1.1f%%', startangle=140)
plt.title("Top Selling Products by Sale Amount")
plt.show()
```

```python
# Analysis 3: Category-wise Sales Trends
category_sales = customer_sales.groupby("Category")["SaleAmount"].sum()
print("\nCategory-wise Sales:\n", category_sales)

# Visualization of Category-wise Sales
plt.figure(figsize=(10, 6))
sns.barplot(x=category_sales.index, y=category_sales.values)
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sale Amount")
plt.xticks(rotation=45)
plt.show()
```

```python
# Analysis 4: Best Days for Sales
sales_by_day = customer_sales.groupby("SaleDay")["SaleAmount"].sum()
print("\nSales by Day:\n", sales_by_day)

# Visualization of Best Days for Sales (Line Plot)
plt.figure(figsize=(10, 6))
days_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
sns.lineplot(x=sales_by_day.index, y=sales_by_day.values, marker="o")
plt.title("Sales by Day of the Week")
plt.xlabel("Day")
plt.ylabel("Total Sale Amount")
plt.xticks(rotation=45)
plt.show()
```

```python
# Analysis 5: Monthly Sales Trends
customer_sales['SaleMonth'] = pd.to_datetime(customer_sales['SaleDate']).dt.strftime("%B")
monthly_sales = customer_sales.groupby("SaleMonth")["SaleAmount"].sum()
monthly_sales = monthly_sales.reindex(["January", "February", "March", "April", "May", "June", 
                                       "July", "August", "September", "October", "November", "December"])
print("\nMonthly Sales:\n", monthly_sales)

# Visualization of Monthly Sales Trends
plt.figure(figsize=(10, 6))
sns.barplot(x=monthly_sales.index, y=monthly_sales.values)
plt.title("Monthly Sales Trends")
plt.xlabel("Month")
plt.ylabel("Total Sale Amount")
plt.xticks(rotation=45)
plt.show()
```
